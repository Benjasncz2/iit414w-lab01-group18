````python
import sys, random
import numpy as np
import warnings          
import pandas as pd              
import matplotlib.pyplot as plt  
import seaborn as sns    
import requests          
import fastf1
import os
import seaborn as sns

RANDOM_SEED = 414  # Course constant. Do not change.
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
warnings.filterwarnings('ignore', category=FutureWarning)

# Environment check
print(f'Python  : {sys.version.split()[0]}')
print(f'NumPy   : {np.__version__}')
print(f'Seed    : {RANDOM_SEED}')
print(f'Pandas  : {pd.__version__}')
print(f'Seaborn : {sns.__version__}')
print(f'Requests : {requests.__version__}')
print(f'FastF1 : {fastf1.__version__}')
````

````python
if not os.path.exists('f1_cache'):
    os.makedirs('f1_cache')
fastf1.Cache.enable_cache('f1_cache')

def get_f1_data(seasons):
    all_results = []
    
    for year in seasons:
        schedule = fastf1.get_event_schedule(year)
        races = schedule[schedule['EventFormat'] != 'testing']
        
        for index, row in races.iterrows():
            gp_name = row['EventName']
            round_num = row['RoundNumber']
            
            try:
                print(f"Descargando: {year} - {gp_name}...")
                session = fastf1.get_session(year, round_num, 'R')
                session.load(laps=False, telemetry=False, weather=False)
                
                results = session.results
                
                df_gp = results[['DriverNumber', 'FullName', 'Abbreviation', 'TeamName', 
                                 'GridPosition', 'Position', 'Status', 'Points']].copy()
                
                df_gp['Season'] = year
                df_gp['Round'] = round_num
                df_gp['GPName'] = gp_name
                df_gp['IsTop10'] = (df_gp['Position'] <= 10).astype(int)
                
                all_results.append(df_gp)
                
            except Exception as e:
                print(f"Error en {gp_name} {year}: {e}")
                
    return pd.concat(all_results, ignore_index=True)

seasons = [2022, 2023, 2024]
df_f1 = get_f1_data(seasons)

# df_f1.to_csv('f1_results_22_24.csv', index=False)

print("\n--- Importación Finalizada ---")
print(f"Total de registros: {len(df_f1)}")
print(df_f1.head())
````

# 4. BASELINE MODEL

## Setup: Create Validation Set

````python
# Create validation set: 2024, Rounds 1-12 (temporal split from partner's EDA)
val_df = df_f1[(df_f1['Season'] == 2024) & (df_f1['Round'] <= 12)].copy()

print(f"Validation set: {len(val_df)} records from {val_df['GPName'].nunique()} races")
print(f"Seasons: {val_df['Season'].unique()}")
print(f"\nTarget distribution:")
print(val_df['IsTop10'].value_counts())
print(f"Proportion Top-10: {val_df['IsTop10'].mean():.2%}")
````

---

# 4.1 DOMAIN HEURISTIC BASELINE

**Rule**: If GridPosition ≤ 10, predict Top-10 finish. Otherwise, predict non-Top-10.

**Why this rule?**
- Partner's EDA showed GridPosition has −0.82 correlation with IsTop10
- Starting positions 1-5 guarantee >90% probability of Top-10 finish
- Position 11 is the critical breakpoint where probability drops below 50%
- This rule encodes the strongest pre-race signal in the data

````python
# Apply baseline rule to validation set
val_df['baseline_prediction'] = (val_df['GridPosition'] <= 10).astype(int)

# Show sample predictions
print("Baseline Rule Application (sample):")
print(val_df[['GPName', 'FullName', 'GridPosition', 'baseline_prediction', 'IsTop10']].head(15).to_string())
````

---

# 4.2 ACCURACY ON VALIDATION SET

**Definition**: Accuracy = (Correct Predictions) / (Total Predictions)

````python
# Calculate accuracy
correct = (val_df['baseline_prediction'] == val_df['IsTop10']).sum()
total = len(val_df)
accuracy = correct / total

print("="*60)
print("BASELINE ACCURACY")
print("="*60)
print(f"Correct predictions: {correct}")
print(f"Total predictions:   {total}")
print(f"\n>>> ACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%) <<<")
print("="*60)
````

---

# 4.3 REFLECTION: IS THIS ACCURACY GOOD ENOUGH?

## Is 80% accuracy good enough to make decisions with?

**Yes, with important caveats.** The 80% represents a real, domain-grounded signal—not statistical luck.

### Evidence of Real Predictive Power:
- A **naive model** that always predicts "Top-10" would score ~50% (since data is perfectly balanced)
- Our **heuristic scores ~80%**, which is a **+30 percentage point improvement**
- This improvement is NOT exploiting class imbalance; it's capturing genuine predictive patterns

### What Could Accuracy Be Hiding?

1. **False Positives**: Drivers we predicted would finish Top-10 but didn't
   - Happens when: Reliability failures (DNF), driver errors, pit stop strategy mistakes
   - Example: Ferrari starts P3, has brake failure, finishes P18

2. **False Negatives**: Drivers we predicted would NOT finish Top-10 but did
   - Happens when: Surprise overtakes, safety cars, weather changes, tire strategy
   - Example: Red Bull starts P15, gains 7 positions through pit strategy, finishes P8

3. **The Majority Class Trap Doesn't Apply Here**:
   - If 50% always finish Top-10, why not always predict "Top-10"? 
   - Answer: That naive baseline scores 50%
   - Our GridPosition heuristic scores 80%
   - **We're definitively better, not just riding the coin flip**

### Practical Implications:
- **80% is useful** but still leaves room for improvement
- **For Lab 2**: This becomes the threshold—any ML model must beat 80% to justify its complexity

````python
# Detailed breakdown: Confusion Matrix
tp = ((val_df['baseline_prediction'] == 1) & (val_df['IsTop10'] == 1)).sum()
fp = ((val_df['baseline_prediction'] == 1) & (val_df['IsTop10'] == 0)).sum()
tn = ((val_df['baseline_prediction'] == 0) & (val_df['IsTop10'] == 0)).sum()
fn = ((val_df['baseline_prediction'] == 0) & (val_df['IsTop10'] == 1)).sum()

print("\nConfusion Matrix Breakdown:")
print(f"  True Positives (TP):  {tp} (predicted Top-10, actually Top-10)")
print(f"  False Positives (FP): {fp} (predicted Top-10, not Top-10)")
print(f"  True Negatives (TN):  {tn} (predicted NOT Top-10, actually not Top-10)")
print(f"  False Negatives (FN): {fn} (predicted NOT Top-10, but actually Top-10)")
print(f"\nAccuracy = (TP + TN) / Total = ({tp} + {tn}) / {total} = {accuracy:.4f}")

# Compare to naive baseline
naive_pred = 1  # Always predict Top-10 (majority class)
naive_acc = (val_df['IsTop10'] == naive_pred).mean()
print(f"\nNaive baseline (always predict Top-10): {naive_acc*100:.2f}%")
print(f"Improvement over naive: +{(accuracy - naive_acc)*100:.2f} percentage points")
````

# 4.5 NO LEAKAGE VERIFICATION

The baseline uses **ONLY pre-race information**. All post-race data is used exclusively for evaluation, never for prediction.

## Feature Audit:

**✓ Pre-Race Features (Safe to use for prediction):**
- `GridPosition`: Determined in qualifying, known BEFORE the race starts
- `Season`, `Round`, `GPName`: Known race schedule
- `DriverNumber`, `FullName`, `TeamName`: Announced driver lineups

**✗ Post-Race Features (LEAKAGE — NOT used for prediction):**
- `Position`: Final race position (only known AFTER race ends)
- `Points`: Awarded after race (only known AFTER race ends)
- `Status`: Determined by race outcome (only known AFTER race ends)

**✓ Target Variable (`IsTop10`):**
- Used **only for EVALUATION** of accuracy
- Derived from `Position` (post-race data)
- **Never** used to make predictions

**✓ Validation Set:**
- 2024 Rounds 1-12 (temporally separated from training data)
- Represents unseen, future data from the model's perspective

## Result:
✅ **No data leakage detected. Baseline is valid.**

## 4.6 Additional Metrics (Stretch)

Beyond accuracy, classification has other metrics:
- **Precision**: Of predicted Top-10, how many were actually Top-10?
- **Recall**: Of actual Top-10 finishes, how many did we predict?
- **F1-Score**: Harmonic mean of Precision and Recall

````python
# Calculate additional metrics from confusion matrix
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print("\nADDITIONAL METRICS:")
print("="*60)
print(f"Accuracy:  {accuracy:.4f} (overall correctness)")
print(f"Precision: {precision:.4f} (reliability of Top-10 predictions)")
print(f"Recall:    {recall:.4f} (coverage of actual Top-10 finishes)")
print(f"F1-Score:  {f1:.4f} (balanced Precision + Recall)")
print("="*60)
````

## 4.7 Second Baseline with sklearn (Stretch)

Compare the heuristic against simple machine learning models to prove domain knowledge is valuable.

````python
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Prepare data
X_val = val_df[['GridPosition']].values
y_val = val_df['IsTop10'].values

# DummyClassifier: Always predict majority class
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(X_val, y_val)
y_pred_dummy = dummy.predict(X_val)

# LogisticRegression: Train on GridPosition
lr = LogisticRegression(random_state=414, max_iter=1000)
lr.fit(X_val, y_val)
y_pred_lr = lr.predict(X_val)

# Compare all three
print("\nMODEL COMPARISON:")
print("="*60)
models = {
    'Heuristic (GridPos ≤ 10)': val_df['baseline_prediction'].values,
    'DummyClassifier (majority)': y_pred_dummy,
    'LogisticRegression (1 feat)': y_pred_lr
}

for name, preds in models.items():
    acc = accuracy_score(y_val, preds)
    prec = precision_score(y_val, preds, zero_division=0)
    rec = recall_score(y_val, preds, zero_division=0)
    f1_s = f1_score(y_val, preds, zero_division=0)
    print(f"{name:35} | Acc: {acc:.4f} | Prec: {prec:.4f} | Rec: {rec:.4f} | F1: {f1_s:.4f}")
print("="*60)
````
